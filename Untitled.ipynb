# Pre-processing

```python
from sklearn.preprocessing import StandardScaler
dataset['normalizedAmount']=StandardScaler().fit_transform(dataset['Amount'].values.reshape(-1,1))
dataset=dataset.drop('Amount',axis=1)
```

```python
dataset.head()
```

```python
dataset = dataset.drop('Time',axis=1)
```

```python
dataset.head()
```

```python
y=dataset.iloc[:,dataset.columns == 'Class']
x=dataset.drop('Class',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test =np.array(y_test)
```

# Deep neural network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units=16,input_dim=29,activation='relu'),
    Dense(units=24,activation='relu'),
    Dropout(0.5),
    Dense(20,activation='relu'),
    Dense(24,activation='relu'),
    Dense(1,activation='sigmoid'),
])
```

```python
model.summary()
```

# Training

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
score = model.evaluate(X_test,y_test)
```

# Under sampling

```python
fraud_indices = np.array(dataset[dataset.Class ==1].index)
#len(y[y.Class ==1])
number_records_fraud = len(fraud_indices)
print(number_records_fraud)
```

```python
normal_indices = dataset[dataset.Class ==0].index
```

```python
random_normal_indices = np.random.choice(normal_indices,number_records_fraud,replace=False)
random_normal_indices = np.array(random_normal_indices)
print(len(random_normal_indices))
```

```python
under_sample_indices = np.concatenate([fraud_indices,random_normal_indices])
under_sample_data=dataset.iloc[under_sample_indices,:]
```

```python
X_undersample = under_sample_data.iloc[:, under_sample_data.columns != 'Class']
y_undersample = under_sample_data.iloc[:, under_sample_data.columns == 'Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_undersample, y_undersample, test_size=0.33)
```

```python
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test =np.array(y_test)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
y_pred = model.predict(X_test)
y_expected=pd.DataFrame(y_test)
cnf_matrix =confusion_matrix(y_expected,y_pred.round())
sns.heatmap(cnf_matrix ,annot =True,fmt='1',xticklabels="P", yticklabels="T" )
```
